```python

import spacy
from spacy.tokens import Doc

class WhitespaceTokenizer:
    def __init__(self, vocab):
        self.vocab = vocab

    def __call__(self, text):
        words = text.strip().split(" ")
        words = [w if w != "" else " " for w in words] 
        #for i,v in enumerate(words):print (f"{i}{v}")
        return Doc(self.vocab, words=words)


```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
nlp.tokenizer = WhitespaceTokenizer(nlp.vocab)
```

```python
import pprint
```

```python
tags = {}
for fyl in fyls:
    fyl.refreshTokens()
    fyl.exfoli8(nlp,show=False)
    for token in fyl.tokens:
        tag = token.mtag
        if tag in [""," ","i","/i"]:
            continue
            
        if tag in tags:
            tags[tag].append(token.text.lower())
        else:
            tags[tag] = [token.text.lower()]

for t in tags.items():
    t[1].sort()
    
pprint.pprint(tags)
```
